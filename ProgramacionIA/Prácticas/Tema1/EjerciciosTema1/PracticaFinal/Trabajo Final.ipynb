El siguiente ejercicio plantea la implementación del clásico juego piedra, papel o tijeras. Se trata de un juego de manos en el que participan dos jugadores. En cada ronda, cada jugador forma, simultáneamente, una de estas tres formas con al mano extendida. Estas formas son i) piedra, que se representa con el puño cerrado, ii) papel, representada con la mano abierta, y iii) tijeras, representada por un puño con los dedos índice y corazón extendidos en forma de V. El juego solo tiene dos resultados posibles. El primero, cuando ambos jugadores realizan la misma acción. El segundo, cuando ambos realizan acciones diferentes. En este caso, la piedra vence (o rompe) a las tijeras, el papel vence (o envuelve) a la piedra y, finalmente, las tijeras vencen (o cortan) el papel.

2. Crea una función llamada *accion_usuario* que solicite al usuario una acción (piedra, papel o tijera) y la devuelva.  

```python
#completar con parámetros de entrada/salida y cuerpo del método
def accion_usuario():
    op1=int(input('Presiona 1 para piedra, 2 para papel, 3 para tijera'))
    if op1>=1 and op1<=3:
        if op1==1:
            op1='piedra'
            print(f'Elegiste {op1}')
        elif op1==2:
            op1='papel'
            print(f'Elegiste {op1}')
        else:
            op1='tijera'
            print(f'Elegiste {op1}')
    else:
        print('Opcion no válida')
        accion_usuario()
    return op1
```

```python
#necesario si quieres utilizar la función descrita en la pista
import random

#completar con parámetros de entrada y cuerpo del método
def accion_ordenador():
    op2=random.randint(1,3)
    if op2==1:
        op2="piedra"
        print(f'Ordenador eligió {op2}')
    elif op2==2:
        op2="papel"
        print(f'Ordenador eligió {op2}')
    else:
        op2="tijera"
        print(f'Ordenador eligió {op2}')
    return op2
```

```python
nombreFichero= input('Dame el nombre del fichero (historal)')
ganar=0
perder=0
empate=0
if nombreFichero == 'historial':
    with open('ficheros/historial.txt', 'r') as fich:
        for linea in fich:
            if 'Tu ganas' in linea:
                ganar=ganar+1
            if 'Tu pierdes' in linea:
                perder=perder+1
            if 'empate' in linea:
                empate=empate+1
    print(f'Ganaste: {ganar}')
    print(f'Perdiste: {perder}')
    print(f'Empates: {empate}')

else:
    print('El fichero no existe')
```

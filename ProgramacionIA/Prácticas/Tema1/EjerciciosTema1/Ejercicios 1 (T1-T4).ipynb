**Ejercicio 1.** Dados dos números ingresados por el usuario devuelve:
* la suma de ellos
* la resta de ellos
* la multiplicación de ambos
* el resultado de la división entera
* el resto de la división

````python
while True:
    num1 = input('Introduce numero 1')
    if(num1.startswith('-') and num1[1:].isdigit()) or num1.isdigit():
        while True:
            num2 = input('Introduce numero 2')
            if(num2.startswith('-') and num2[1:].isdigit()) or num2.isdigit():
                break
            else:
                print(f'{num2} no es un numero válido.')
        break
    else:
        print(f'{num1} no es un numero válido.')

num1= int(num1)
num2= int(num2)

print(f'La suma de {num1} y {num2} es {num1+num2}')
print(f'La resta de {num1} y {num2} es {num1-num2}')
print(f'La multiplicación de {num1} y {num2} es {num1*num2}')
if(num2==0):
    print(f'No se puede dividir entre {num2}')
    print(f'No se puede sacar el módulo de {num2}')
else:
    print(f'La división de {num1} y {num2} es {num1/num2}')
    print(f'El módulo de {num1} y {num2} es {num1%num2}')
````

**Ejercicio 23.** Verifica si una cadena ingresada por el usuario es un pangrama (contiene todas las letras del alfabeto).

````python
cadena= input('Introduce una cadena: ')
cadena = cadena.replace(" ", "")
abecedario='abcdefghijklmnopqrstuvwxyzñ'
arrlistaCaracteres = list(cadena)
arrlistaCaracteresOrdenada = sorted(arrlistaCaracteres)
listaOrdenada = ''.join(arrlistaCaracteresOrdenada)

if(listaOrdenada == abecedario):
    print('La cadena es un pentagrama')
else:
    print('La cadena no es un pentagrama')
````

**Ejercicio 24.** Elimina las vocales de una cadena ingresada por el usuario.

**Ejercicio 26.** Solicita al usuario dos cadenas, (1) una frase y (2) un acrónimo. El programa indicará si el acrónimo corresponde a la frase indicada.

OJO. La primera cadena que introduzca el usuario deberá contener en mayúsculas aquellas letras que pertenecen al acronimo.

````python
frase= input('Introduce una frase: ')
acronimo= input('Introduce su acronimo: ')
arrlistaCaracteres = frase.split(" ")
esAcronimo = True

for i in range(len(arrlistaCaracteres)):
    if arrlistaCaracteres[i][0].lower() == acronimo[i].lower():
        continue
    else:
        print('La frase no es un acronimo')
        esAcronimo = False
        break

if esAcronimo:
    print('La frase es un acronimo')
else:
    print('La frase no es un acronimo')
````

**Ejercicio 1.** Dados dos números ingresados por el usuario devuelve:
* la suma de ellos
* la resta de ellos
* la multiplicación de ambos
* el resultado de la división entera
* el resto de la división

````python
num1 = input('Introduce numero 1')
num2 = input('Introduce numero 2')

print(f'La suma de {num1} y {num2} es {int(num1)+int(num2)}')
print(f'La resta de {num1} y {num2} es {int(num1)-int(num2)}')
print(f'La multiplicación de {num1} y {num2} es {int(num1)*int(num2)}')
print(f'La división de {num1} y {num2} es {int(num1)/int(num2)}')
print(f'El módulo de {num1} y {num2} es {int(num1)%int(num2)}')
````

**Ejercicio 2.** Calcula el valor absoluto de un número entero ingresado por el usuario.

````python
print(f'El valor absoluto de {num1} y {num2} son {abs(int(num1))} y {abs(int(num2))}')
````

**Ejercicio 3.** Verifica si un número entero ingresado por el usuario es par o impar.

````python
if(int(num1)%2 == 0):
    print('num1 es par')
else:
    print('num1 es impar')
if(int(num2)%2 == 0):
    print('num2 es par')
else:
    print('num2 es impar')
````

**Ejercicio 4.** Verifica si un número entero ingresado por el usuario es positivo, negativo o cero.

````python
if(int(num1) > 0):
    print('num1 es positivo')
else:
    if(int(num1) < 0):
        print('num1 es negativo')
    else:
        print('num1 es cero')

if(int(num2) > 0):
    print('num2 es positivo')
else:
    if(int(num2) < 0):
        print('num2 es negativo')
    else:
        print('num2 es cero')
````

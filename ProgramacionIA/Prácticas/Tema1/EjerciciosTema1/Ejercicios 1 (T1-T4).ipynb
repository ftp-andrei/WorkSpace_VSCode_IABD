**Ejercicio 1.** Dados dos números ingresados por el usuario devuelve:
* la suma de ellos
* la resta de ellos
* la multiplicación de ambos
* el resultado de la división entera
* el resto de la división

````python
while True:
    num1 = input('Introduce numero 1')
    if(num1.startswith('-') and num1[1:].isdigit()) or num1.isdigit():
        while True:
            num2 = input('Introduce numero 2')
            if(num2.startswith('-') and num2[1:].isdigit()) or num2.isdigit():
                break
            else:
                print(f'{num2} no es un numero válido.')
        break
    else:
        print(f'{num1} no es un numero válido.')

num1= int(num1)
num2= int(num2)

print(f'La suma de {num1} y {num2} es {num1+num2}')
print(f'La resta de {num1} y {num2} es {num1-num2}')
print(f'La multiplicación de {num1} y {num2} es {num1*num2}')
if(num2==0):
    print(f'No se puede dividir entre {num2}')
    print(f'No se puede sacar el módulo de {num2}')
else:
    print(f'La división de {num1} y {num2} es {num1/num2}')
    print(f'El módulo de {num1} y {num2} es {num1%num2}')
````

**Ejercicio 2.** Calcula el valor absoluto de un número entero ingresado por el usuario.

````python
while True:
    num1 = input('Introduce numero')
    if(num1.startswith('-') and num1[1:].isdigit()) or num1.isdigit():
        num1 = int(num1)
        break
    else:
        print('El valor introducido no es un numero')
print(f'El valor absoluto de {num1} es {abs(num1)}')
````

**Ejercicio 3.** Verifica si un número entero ingresado por el usuario es par o impar.

````python
num1 = input('Introduce numero')
if (num1.isdigit() and int(num1) > 0):
    num1= int(num1)
    if(num1%2 == 0):
        print('num1 es par')
    else:
        print('num1 es impar')
else:
    print('El numero introducido no es correcto o es 0')
````

**Ejercicio 4.** Verifica si un número entero ingresado por el usuario es positivo, negativo o cero.

````python
num1 = input('Introduce numero')
if(num1.startswith('-') and num1[1:].isdigit()) or num1.isdigit():
    num1 = int(num1)
    if(num1 > 0):
        print('num1 es positivo')
    elif(num1 < 0):
        print('num1 es negativo')
    else:
        print('num1 es cero')
else:
    print('Eso no es un numero')
    
````

**Ejercicio 6.** Calcula el factorial de un número entero ingresado por el usuario.

````python
num1 = int(input('Introduce el número: '))
suma = 1
for i in range(1, num1 + 1):
    suma = suma*i
print(f'El valor de la suma de {num1} es {suma}')

````

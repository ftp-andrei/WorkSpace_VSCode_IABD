**Ejercicio 9**. Solicita el nombre de un fichero al usuario. Si el fichero no existe, lo creará. Posteriormente, solicitar al usuario una frase hasta que introduzca 'finalizar', las frases se irán añadiendo al final del fichero.

````python
nombreFichero= input('Dame nombre del fichero')
fich = open(f'ficheros/{nombreFichero}.txt', 'a+')
finalizar=True

while finalizar:
    frase= input('Dame una frase o escribe finalizar para terminar')
    if frase.lower() != 'finalizar':
        fich.write(frase+'\n')
    else:
        finalizar = False
fich.close()
````

**Ejercicio 10.a**. Crea una función llamada `procesar_fichero(nombreFichero)` que devuelva un diccionario que represente un registro del número de apariciones de cada palabra en el fichero. El parámetro de entrada será el nombre del fichero.

````python
def procesar_fichero(nombreFichero):
    diccionario = {}
    fich = open(nombreFichero, "r")
    for linea in fich:
        linea = linea.strip()  
        palabras = linea.split(" ")
        for pal in palabras:
            if pal in diccionario:
                diccionario[pal] += 1
            else:
                diccionario[pal] = 1
    fich.close()
    return diccionario

nombreFichero = input('Introduce el nombre del fichero: ')
procesar_fichero(f'ficheros/{nombreFichero}.txt')
````
